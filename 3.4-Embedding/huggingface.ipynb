Embedding Techniques  Using HuggingFace 

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HF_TOKEN']=os.getenv('HF_LANGCHAIN_TOKEN')

# here in huggingface we need to pip install sentence-transformers which will be having embedding techniques
# we will also install langchain_huggingface which will have many techniques of calling LLM models


# sentence transformers on hugging face
#  Huggingface sentence transformers is python framework for state-of-the-art sentence, text and image embeddings. one of the embedding models is used in bhuggingfaceembeddings class. we also have added an alis=as for SentenceTransformerEmbedding for users who are more familiar with directly using that package

from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
text="this is a test document"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_results=embeddings.embed_documents([text,"its not a test document"])
doc_results
```
