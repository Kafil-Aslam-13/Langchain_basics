### Ollama

ollama supports embedding models , making it possible to build Retrieval Agumented Generation (RAG) Applications that combine text prompts with existing documents or other data 

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma2:2b")  # by default it uses lama 2
)
```

```python
## creating vectors
r1=embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet"
    ]
)
```

```python
r1
```

```python
len(r1[0])
```

```python
embeddings.embed_query("Beta is second letter of Greek alphabet")
```

```python
# olama has different diffetrent embedding models check https://ollama.com/blog/embedding-models 
embeddings=OllamaEmbeddings(model="mxbai-embed-large")
text="This is a test documnent "
query_result=embeddings.embed_query(text)
query_result
```
